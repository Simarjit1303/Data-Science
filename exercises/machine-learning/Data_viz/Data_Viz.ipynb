```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
birds_data = pd.read_csv('https://raw.githubusercontent.com/microsoft/Data-Science-For-Beginners/refs/heads/main/data/birds.csv')
birds_data.head()
```

```python
plt.style.use('seaborn-v0_8-bright')
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-', alpha = 0.5)
plt.plot(x, np.cos(x), '--')

fig.savefig('figure1.png')
```

### Object oriented interface

```python
fig, ax = plt.subplots(2)
print(f"type of fig:\n{type(fig)},\n\ntype of ax:\n{type(ax)}\n")

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

### Adjusting the plot


```python
plt.plot(x, np.sin(x - 0), color = 'blue')
plt.plot(x, np.sin(x - 1), color = 'g')
plt.plot(x, np.sin(x - 2), color = '0.75')
plt.plot(x, np.sin(x - 3), color = '#FFDD44')

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x - 0), color = 'blue')
plt.plot(x, np.sin(x - 1), color = 'g')
plt.plot(x, np.sin(x - 2), color = '0.75')
plt.plot(x, np.sin(x - 3), color = '#FFDD44')

plt.xlim(0, np.pi * 2)
plt.ylim(-1.5, 1.5)
```

### Adding legends

```python
plt.plot(x, np.sin(x), '-g',label = 'sin(x)')
plt.plot(x, np.cos(x), ':b',label = 'cos(x)')

plt.title("Some curves")
plt.xlabel("x")
plt.ylabel("sin(x) or cos(x)")

plt.legend(loc = 'best')
```

```python
fig, ax = plt.subplots(1)

ax.plot(x, np.sin(x), color = 'green',label = 'sin(x)')
ax.plot(x, np.cos(x), color = 'blue',label = 'cos(x)')

ax.set_title("Some curves")
ax.set_xlabel("x")
ax.set_xlim(0, np.pi * 2)
ax.set_ylabel("sin(x) or cos(x)")


ax.legend(loc = 'best')
ax.get_xlim()
```

```python
fig, ax = plt.subplots(1)

x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
ax.plot(x, np.sin(x), label='sin(x)')
ax.plot(x, np.cos(x), label='cos(x)')
ax.set_title("Sine and Cosine curves")
ax.set_xlabel("x")
ax.set_ylabel("sin(x) or cos(x)")
ax.legend()
```

### Scatter plots

```python
fig, ax = plt.subplots(1)
fig.set_size_inches(4, 3)
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
sizes = np.linspace(1, 200, 1000)
ax.scatter(x, np.sin(x), alpha = .1, s=sizes)
# ax.set_xlim(-np.pi,np.pi)
```
