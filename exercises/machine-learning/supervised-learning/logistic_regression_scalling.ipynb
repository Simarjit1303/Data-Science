```python
import pandas as pd
from sklearn import model_selection, preprocessing, linear_model, metrics
```

# Data Collection

```python
df = pd.read_csv('https://raw.githubusercontent.com/m-mahdavi/teaching/refs/heads/main/datasets/mnist.csv')
df.head()
```

```python
df.describe()
```

```python
# spliting the data into 2 parts with 25% for testing and 75% for training
df_train, df_test = model_selection.train_test_split(df)
print(f"df_size: {df.shape}")
print(f"df_train size: {df_train.shape}")
print(f"df_test size: {df_test.shape}")
```

# Data Reprocessing

```python
# removing the explicit data from the dataset
df_train = df_train.dropna()
df_test = df_test.dropna()

# printing the result to check the new values in a datset
print(f"X_train_size: {df_train.shape}")
print(f"X_test_size: {df_test.shape}")
```

```python
x_train = df_train.drop('class', axis=1)
y_train = df_train['class']
x_test = df_test.drop('class', axis=1)
y_test = df_test['class']

print("x_train size:",x_train.shape)
print("y_train size:",y_train.shape)
print("x_test size:",x_test.shape)
print("y_test size:",y_test.shape)
```

```python
# Scaling the dataset
scaler = preprocessing.StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# Data Modelling

```python
model = linear_model.LogisticRegression()
model.fit(x_train_scaled, y_train)
```

```python
y_predict = model.predict(x_test_scaled)
```

```python
accuracy =  metrics.accuracy_score(y_test, y_predict)
print(f"accuarcy of model is : {accuracy}")
```
